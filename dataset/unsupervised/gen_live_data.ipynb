```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("UNSW-NB15_1.csv")
data
```

```python
value_counts = data["label"].value_counts()
count_0 = data['label'].value_counts().get(0, 0)
count_1 = data['label'].value_counts().get(1, 0)
total_samples = len(data)
count_1 / total_samples
```

```python
desired_count_0 = int(100000 * count_0 / total_samples)
desired_count_1 = int(100000 * count_1 / total_samples)

data_0 = data[data['label'] == 0].sample(n=desired_count_0, replace=True)
data_1 = data[data['label'] == 1].sample(n=desired_count_1, replace=True)

filtered_data = pd.concat([data_0, data_1])

filtered_data = filtered_data.reset_index(drop=True)

filtered_data
```

```python
filtered_data.to_csv('unsupervised_data.csv', index=False)
```
